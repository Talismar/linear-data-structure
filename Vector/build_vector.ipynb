```python
# Answers - 1 and 2

class UnorderedArray:

  insert_index = 0

  def __init__(self, lenght):
    self.dim = lenght
    self.vector = [None] * lenght
  
  # Inserir
  def insert(self, new_value):
    if self.insert_index < self.dim: 
      self.vector[self.insert_index] = new_value
      self.insert_index += 1
    else:
      print("Index limit")

  # Apagar
  def remove(self, value):
    index = self.linearSearch(value)
  
    if type(index) == str:
      return "Not found!!!"

    self.vector[index] = None
    
    for i in range(index, self.dim):
      if i >= index: 
        try:
          self.vector[i], self.vector[i+1] = self.vector[i+1], None
        except IndexError:
          pass

    return True

  # BuscarLinar
  def linearSearch(self, value):
    for index, i in enumerate(self.vector):
      if i == value:
        return index 

    return "Not found"

  # Print Vetor
  def __str__(self):
    ret = '['

    for i in self.vector:
      if i != None:
        ret += str(i) + ', '
    
    return ret[:-2] + ']'
```

```python
# Answers - 3, 4 and 5
import time
import numpy as np
import pandas as pd

class OrderedArray:
  def __init__(self, lenght = 0):
    self.it = 0
    self.dim = lenght
    self.vector = [None] * lenght

  # Inserir    
  def insert(self, value):
    if self.it < len(self.vector):
      
      if self.it == 0: # First insert
        self.vector[self.it] = value
        self.it += 1

      elif value >= self.vector[self.it - 1]: 
        self.vector[self.it] = value
        self.it += 1

      else:
        for (i, j) in enumerate(self.vector):
          if value < j:
            # print(value, i, j, self.it)

            for k in range(self.it ,i, -1 ):
              self.vector[k], self.vector[k-1] = self.vector[k-1], None
            
            self.vector[i] = value
            self.it += 1
            break
  
   # Apagar
  def remove(self, value):
    index = self.Linear_search(value)
  
    if type(index) == str:
      return "Not found!!!"

    self.vector[index] = None
    
    for i in range(index, self.dim):
      if i >= index: 
        try:
          self.vector[i], self.vector[i+1] = self.vector[i+1], None
        except IndexError:
          pass

    return True

  # BuscarLinear            
  def Linear_search(self, value):
    for index, i in enumerate(self.vector):
      if i == value:
        return index 

    return "Not found"

  # BuscarBinária
  def binary_search(self, value):
    start = 0
    end = self.it
    index = end // 2 

    if value > self.vector[end-1] or value < self.vector[start]:
      return "Not found"

    while value != self.vector[index]:      
      # print("Middle:",index)
      
      if start > end:
        return "--|Not Found|--"

      elif value < self.vector[index]:
        end = index - 1
        # print("<", start, "+", end, "//", 2)
        
        index = (start + end) // 2

      elif value > self.vector[index]:
        start = index + 1
        # print(">", start // end, "//", 2)

        index = (start + end) // 2

    return index, "|", value

  #ImprimirValor
  def __str__(self) -> str:
    return str(self.vector)
```

```python
def question_05():
  SIZE_ = [10,100,1000,2000,4000,8000,16000,32000]
  dict_ret = {"-": [],"N elementos": [], "Busca Linear": [], "Busca Binária": []}

  for i in SIZE_:
    v = OrderedArray(i)
    dict_ret["N elementos"].append(i)

    for j in np.floor(np.random.uniform(0, 100, i)):
      v.insert(int(j))

    dict_ret["-"].append(f"{i//3} - {v.vector[i//3]}")

    startTime_Linear = time.time()
    v.Linear_search(v.vector[i//3])
    endTime_Linear = time.time()

    dict_ret["Busca Linear"].append(endTime_Linear - startTime_Linear)

    startTime_Binary = time.time()
    v.binary_search(v.vector[i//3])
    endTime_Binary = time.time()

    dict_ret["Busca Binária"].append(endTime_Binary - startTime_Binary)

  return dict_ret

v = OrderedArray(5)

for i in [1,3,5,7,9]:
  v.insert(i)

# v.remove(3)

print(v.vector)

# print(v.binary_search(1))
```
